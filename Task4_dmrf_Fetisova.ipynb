```python
# Создаем стобец с типом норматива
def norm_type(x):
  if str(x)[0] == 'в':
    return 'в пределах'
  else:
    return 'не более'

df['Тип норматива'] = df['Норматив'].apply(norm_type)
df
```

```python
df['Норматив']=df['Норматив'].replace('в пределах ','',regex=True)
df['Норматив']=df['Норматив'].replace('не более ','',regex=True)
df['Норматив']=df['Норматив'].replace(',','.',regex=True)
df
```

```python
# Создаем стобцы с нормативами
def left_n(x):
  l=x.split("-")
  lef=l[0]
  return lef
df['Норматив_нижняя граница'] = df['Норматив'].apply(left_n).astype(float)
def right_n(x):
  l=x.split("-")
  righ=l[-1]
  return righ
df['Норматив_верхняя граница'] = df['Норматив'].apply(right_n).astype(float)
#df['Норматив_верхняя граница'] = df['Норматив'].apply(right_n).astype(float)
df
```

```python
# находим в столбце с результом неверные значения, заменяем их на -9999999999, весь столбец меняем на float
def error_type(x):
  try:
        float(x)
        return str(x)
  except ValueError:
        return '-99999999999'

df['Результат анализа'] = df['Результат анализа'].apply(error_type)
df['Результат анализа'] = df['Результат анализа'].astype(float)
df
```

```python
# Создаем стобец с результатом
pd.options.mode.chained_assignment = None
df['Результат'] = ''
df['Результат'][(df['Тип норматива'] == 'В пределах') & (df['Результат анализа'] <= df['Норматив_верхняя граница']) & (df['Результат анализа'] >= df['Норматив_нижняя граница'])] = 'Норма'
df['Результат'][(df['Тип норматива'] == 'В пределах') & (df['Результат анализа'] > df['Норматив_верхняя граница']) | (df['Результат анализа'] < df['Норматив_нижняя граница'])] = 'Вне нормы'
df['Результат'][(df['Тип норматива'] == 'не более') & df['Результат анализа'] <= df['Норматив_верхняя граница']] = 'Норма'
df['Результат'][(df['Тип норматива'] == 'не более') & df['Результат анализа'] > df['Норматив_верхняя граница']] = 'Вне нормы'
df['Результат'][df['Результат анализа'] == -99999999999] = 'некорректные данные'
df
```
