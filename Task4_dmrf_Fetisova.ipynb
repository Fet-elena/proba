```python
# Создаем стобец с типом норматива
def norm_type(x):
  if str(x)[0] == 'в':
    return 'в пределах'
  else:
    return 'не более'

df['Тип норматива'] = df['Норматив'].apply(norm_type)
df
```

```python
df['Норматив']=df['Норматив'].replace('в пределах ','',regex=True)
df['Норматив']=df['Норматив'].replace('не более ','',regex=True)
df['Норматив']=df['Норматив'].replace(',','.',regex=True)
df
```

```python
# Создаем стобцы с нормативами
def left_n(x):
  l=x.split("-")
  lef=l[0]
  return lef
df['Норматив_нижняя граница'] = df['Норматив'][(df['Тип норматива'] == 'в пределах')].apply(left_n).astype(float)
df['Норматив_нижняя граница'] = df['Норматив_нижняя граница'].fillna(0)
def right_n(x):
  l=x.split("-")
  righ=l[-1]
  return righ
df['Норматив_верхняя граница'] = df['Норматив'].apply(right_n).astype(float)
df
```

```python
# находим в столбце с результом строковые значения, заменяем их на 0, весь столбец меняем на float
def error_type(x):
  try:
        float(x)
        return str(x)
  except ValueError:
        return '0'

df['Результат анализа'] = df['Результат анализа'].apply(error_type)
df['Результат анализа'] = df['Результат анализа'].astype(float)
df
```

```python
# Создаем стобец с результатом
df['Результат'] = np.where((df['Результат анализа'] <= df['Норматив_верхняя граница']) & (df['Результат анализа'] >= df['Норматив_нижняя граница']), 'Норма', 'Вне нормы')
df
```

**Задача 2** В ящике 5 апельсинов и 4 яблока. Наудачу выбираются 3 фрукта. Какова вероятность, что все три фрукта – апельсины?

В интернете полученный аналитически ответ 0.119. Подтверждается ли он эксперементально?

```python
B1 = ["apple"]*4 + ['orange']*5
# задаем число экспериментов
n = 1000000
# # формируем выборку
A = pd.DataFrame({"A":list(map(lambda a: random.sample(B1, 3, counts=None), range(n)))})
# # считаем
P = len(A[A.A.astype('str') =="['orange', 'orange', 'orange']"])/A.shape[0]
P
```

**Задача 3** Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. Какова вероятность, что он проверит ровно две детали?

```python
D = ["stand"]*7 + ['nestand']*3
# задаем число экспериментов
n = 1000000
# # формируем выборку
C = pd.DataFrame({"C":list(map(lambda a: random.sample(D, 2, counts=None), range(n)))})
# # считаем
R = len(C[C.C.astype('str') =="['nestand', 'stand']"])/C.shape[0]
R
```
