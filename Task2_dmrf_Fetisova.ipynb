```python
import pickle as pkl
import pandas as pd
with open("stroj.pkl", "rb") as f:
    object = pkl.load(f)
df = pd.DataFrame(object)
df
```

```python
df.index = np.arange(df.shape[0]) #индексируем от 0 до количества строк в таблице

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.isnull().sum())   # вывод всех столбцов с количеством в них пропусков
```

```python
df.drop(columns=['objPriceAVG', 'objCommercNm', 'complexShortName', 'problemFlag'], axis= 1 , inplace= True) # удаляем столбцы с большим количеством пропусков
# df.count(axis='columns')
df
```

```python
# количество пропусков
df.isna().sum()
```

```python
# смотрим инфу
df.info()
```

```python
# посмотрим min, max, среднее и тд.
df.describe().T
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
# Рассчитываем средние цены за квадратный метр по регионам
plt.rcParams['figure.figsize'] = (15, 10)
df.groupby('rpdRegionCd').objSquareLiving.mean().plot(kind='bar', color='darkgreen')
plt.title("Стоимость квадратного метра по регионам", color='g', fontsize=16)
plt.ylabel('Стоимость за кв.м., руб.', fontsize=14, color='orange')
plt.xlabel('Регион', fontsize=14, color='orange')
#fig = px.bar(df,x = "Students",y = "Age",color = "Students")
#fig.update_layout(xaxis = {"categoryorder":"total ascending"})

# делаем вывод, что самые дорогие квартиры в Москве и Чечне
```

```python
# Рассчитываем объемы вводимого жилья по регионам и по годам
df.groupby(['rpdRegionCd']).objSquareLiving.count().plot(kind='barh', color='darkgreen')
plt.rcParams['figure.figsize'] = (10, 15)
# df.groupby(['rpdRegionCd']).objSquareLiving.count().plot(kind='pie')
# plt.pie('rpdRegionCd', labels=labels)
plt.title("Количество строящихся объектов по регионам", color='g', fontsize=16)
plt.ylabel('Регион', fontsize=14, color='orange')
plt.xlabel('Количество', fontsize=14, color='orange')

# Делаем вывод, что больше всего строят в Краснодарском крае
```

```python
# Рассчитываем объемы вводимого жилья по годам
df['year'] = df['objReady100PercDt'].apply(lambda x: x.split('-')[0])
df.groupby(['year']).objSquareLiving.count().plot(kind='pie')
plt.title("Количество строящихся объектов по годам")
```

```python
df.groupby(['rpdRegionCd'], as_index = False).objSquareLiving.mean().sort_values(by = 'objSquareLiving')
```
