```python
labeled = pd.read_csv("labeled.csv")
labeled
```

```python
from typing_extensions import dataclass_transform
comment=[]
for i in labeled.comment:
  comment.append(i)
comment[:10]
```

```python
tokenizer = WordPunctTokenizer()
data_tok = []
for i in comment:
  data_tok.append(tokenizer.tokenize(i.lower()))
data_tok[:1]
```

```python
import warnings
warnings.filterwarnings("ignore")

def vectorize(tokens):
    vector=[]
    for w in filtered_vocab:
        vector.append(tokens.count(w))
    return vector

def unique(sequence):
    seen = set()
    return [x for x in sequence if not (x in seen or seen.add(x))]

nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))
special_char=[",",":"," ",";",".","?"]

snowball = SnowballStemmer(language="russian")
morph = MorphAnalyzer()

data_tok_lem_stem = []
for i in data_tok:
    string = []
    for j in i:
        word = snowball.stem(morph.normal_forms(j)[0])
        string.append(word)
    data_tok_lem_stem.append(string)

lem_stem_words = []
for i in data_tok_lem_stem:
  for j in i:
    lem_stem_words.append(j)

vocab=unique(lem_stem_words)

filtered_vocab=[]
for w in vocab:
    if w not in stop_words and w not in special_char:
        filtered_vocab.append(w)

vector = []
for i in data_tok_lem_stem:
    vector.append(vectorize(i))
print('Количество уникальных слов:',len(vocab))
print('Количество уникальных слов без мусора:',len(filtered_vocab))
print('Количество векторов:',len(vector))
```
