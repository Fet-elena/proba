```python
import requests
```

## Шаг 1й - Получение списка объектов

```python
offset_ = 1 # размер шага
limit_ = 100 # количество объектов в выдаче
baza = []
while True:
  url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
  res = requests.get(url)
  objects_data = res.json()
  baza.extend(objects_data.get('data').get('list'))
  offset_ = offset_ + limit_
  if len(objects_data.get('data').get('list')) < 100:
    break
a = len(baza)
a


```

Загрузилось 11655 объектов

## Шаг 2 - преобразование в Датафрейм

```python
import pandas as pd
df = pd.DataFrame(baza)
df_stroj = df[df.siteStatus == 'Строящиеся']
df_stroj
```
